```python
import numpy as np
```

```python
def sigmoid(x):
    result = 1/(1 + 1/np.power(np.e, x))
    return result
```

```python
assert sigmoid(0) == 0.5
assert sigmoid(6) > 0.99
assert sigmoid(np.inf) == 1
assert sigmoid(np.NINF) == 0
```
